```python
import pandas as pd

# Load your spreadsheet
xlsx = pd.ExcelFile('output.xlsx')

# Load a sheet into a DataFrame by name
df = pd.read_excel(xlsx, 'Sheet1')

# Strip white space and trailing " -" from the first column and make comparison case-insensitive
df[df.columns[0]] = df[df.columns[0]].str.strip().str.lower().str.rstrip(' -')

# Remove the duplicates
df_no_duplicates = df.drop_duplicates(subset=df.columns[0])

# Save DataFrame to .xlsx file
df_no_duplicates.to_excel('refined_file.xlsx', index=False)
```

```python
# Load the Excel file
df = pd.read_excel('Hospital List.xlsx')

# Drop the rows where at least one element is missing
df = df.dropna()

# Convert 'coordinates' from string to dictionary
df['coordinates'] = df['coordinates'].apply(ast.literal_eval)

# Extract 'lat' and 'lng' into separate columns
df['latitude'] = df['coordinates'].apply(lambda x: x['lat'])
df['longitude'] = df['coordinates'].apply(lambda x: x['lng'])

# Create a map centered around the first coordinate
m = folium.Map(location=[df['latitude'].iloc[0], df['longitude'].iloc[0]], zoom_start=13)

# Add points to the map
for index, row in df.iterrows():
    folium.Marker([row['latitude'], row['longitude']], popup=row['name']).add_to(m)

# Display the map
m

```
